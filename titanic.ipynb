```python
df.dropna(subset = ["Embarked"], inplace = True, axis = 0)

df.reset_index(drop = True, inplace = True)
```

```python
avg_age = df["Age"].mean()
print(round(avg_age,0))
```

```python
df["Age"].replace(np.nan, round(avg_age,0), inplace = True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=df)
plt.title('Distribution of Survival')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', data=df)
plt.title('Distribution of Passenger Classes')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'].dropna(), kde=True)
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df[df['Survived'] == 1]['Age'].dropna(), kde=True, color='green', label='Survived')
sns.histplot(df[df['Survived'] == 0]['Age'].dropna(), kde=True, color='red', label='Not Survived')
plt.legend()
plt.title('Survival Rate by Age')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
lableencoder = LabelEncoder()

df["Sex"] = lableencoder.fit_transform(df["Sex"])

df.head()
```

```python
df["Sex"], df["Survived"]
```

```python
X = df[['Pclass', 'Sex']]
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression

log = LogisticRegression(random_state = 0)
log.fit(X_train, y_train)
```

```python
pred = print(log.predict(X_test))
```
